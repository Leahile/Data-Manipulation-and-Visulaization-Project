```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]


plt.plot(x, y, marker='o', linestyle = '-')

for i, (xi, yi) in enumerate(zip(x, y)):
    plt.annotate(f'({xi}, {yi})', (xi,yi), textcoords='offset points', xytext=(0,10), ha= 'center')

plt.title("Line chart with annotations")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x*2

plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Fist Graph")
plt.show()

plt.figure()
x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]
plt.plot(x1, y1, '-.')
plt.title("Second Graph")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x*2

plt.plot(x, y)

x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]

plt.plot(x1, y1, '-.')
plt.title("Multiple plots graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x*2

plt.plot(x, y)

x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]

plt.plot(x, y1, '-.')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Plots")
plt.fill_between(x, y, y1, color='yellow', alpha=0.5)
plt.show()
```

```python
#Bar Graph
import matplotlib.pyplot as plt
import numpy as np

fruits = ["Banana", "Cherries", "Apples", "Grapes"]
sales = [400, 350, 300, 450]

plt.bar(fruits, sales, color='violet') #customising bars
plt.title("Fruit Sales")
plt.xlabel("Fruits")
plt.ylabel("Sales")
plt.show()
```

```python
#Adjusting Bars 
import matplotlib.pyplot as plt
import numpy as np

fruits = ["Banana", "Cherries", "Apples", "Grapes"]
sales = [400, 350, 300, 450]

plt.bar(fruits, sales, color='yellow', width=0.3) #customising bars
plt.title("Fruit Sales")
plt.xlabel("Fruits")
plt.ylabel("Sales")
plt.show()
```

```python
#Bar Graph
import matplotlib.pyplot as plt
import numpy as np

fruits = ["Banana", "Cherries", "Apples", "Grapes"]
sales = [400, 350, 300, 450]

plt.barh(fruits, sales, color='violet') #horizontal bars
plt.title("Fruit Sales")
plt.xlabel("Fruits")
plt.ylabel("Sales")
plt.show()
```
