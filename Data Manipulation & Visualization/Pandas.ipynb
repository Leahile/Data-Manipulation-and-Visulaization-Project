Filling Missing Values in Pandas


```python
#Using fillna()
d = {'First Score': [100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score': [np.nan, 40, 80, 98]}
df = pd.DataFrame(d)

print("Oridinal Dataframe:\n", df)
print("New Dataframe:\n", df.fillna(0))
print("\n")
forward_fill =df.fillna( method='pad')
backward_fill = df.fillna( method='bfill')

print("Filling missing values with the previous value:\n", forward_fill)
print("Filling missing values with the next value:\n", backward_fill )
```
