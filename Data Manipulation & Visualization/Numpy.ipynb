```python
#filtering
ages = np.array([[18, 24, 17, 32, 16, 20, 19],
                 [19, 18, 15, 29, 17, 19, 18]])

teenagers = ages[ages < 18]
print(teenagers)

young_adults = ages[(ages >= 18) & (ages <29)]
print(young_adults)

seniors = ages[ages >29]
print(seniors)

adults = np.where(ages >=18, ages, 0)
print(adults)
```

```python
#Appending values at the end of an NumPy array
#Appending a Single Value to a 1D Array
arr = np.array([1, 8, 3, 3, 5])
print('Original Array : ', arr)

# appending to the array
arr = np.append(arr, [6, 7, 8, 9, 10])
print('Array after appending : ', arr)

#appending another array at the end of a 1D array
arr1 = np.array([1, 2, 3])
print('First array is : ', arr1)

# creating another array
arr2 = np.array([4, 5, 6])
print('Second array is : ', arr2)

arr3 = np.array([7, 8, 9])
print('Third array is:', arr3)

# appending arr2 and arr3 to arr1
arr = np.append(arr1, [arr2, arr3])
print('Array after appending : ', arr)

#Appending Values at the End Using Concatenation
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6]])

combined = np.concatenate((arr1, arr2), axis=0)
print(combined)

arr3 = np.array([2, 4, 6])
arr4 = np.array([3, 6, 9])

combined2 = np.concatenate((arr3, arr4), axis=0)
print(combined2)

#Appending with a Different Array Type
# importing the module
arr = np.array([1, 2, 3])
arr_float = np.array([4.0, 5.0])

combined = np.append(arr, arr_float)
print(combined)  
print("\n")
#Appending Values at the End of the N-Dimensional Array
arr = np.arange(1, 13).reshape(2, 6)
print('Original Array')
print(arr, '\n')
# create another array which is
# to be appended column-wise
col = np.arange(5, 11).reshape(1, 6)
print('Array to be appended column wise')
print(col)
arr_col = np.append(arr, col, axis=0)
print('Array after appending the values column wise')
print(arr_col, '\n')

# create an array which is
# to be appended row wise
row = np.array([1, 2]).reshape(2, 1)
print('Array to be appended row wise')
print(row)
arr_row = np.append(arr, row, axis=1)
print('Array after appending the values row wise')
print(arr_row)
```

```python
#Example: Vertical Stacking of 1D Arrays Using numpy.vstack()
print("Horizontal Stack")
a = np.array([ 1, 2, 3] )
print ("1st Input array : \n", a) 

b = np.array([ 4, 5, 6] )
print ("2nd Input array : \n", b) 

res = np.vstack((a, b))
print ("Output vertically stacked array:\n ", res)


print("\nVertical Stack")

#Vertical Stacking of 2D Arrays Using numpy.vstack()
a = np.array([[ 1, 2, 3], [ -1, -2, -3]] )
print ("1st Input array : \n", a) 

b = np.array([[ 4, 5, 6], [ -4, -5, -6]] )
print ("2nd Input array : \n", b) 

res = np.vstack((a, b))
print ("Output stacked array :\n ", res)
```

```python
#How to swap columns of a given NumPy array?
#Using indexing
a = np.arange(12).reshape(4, 3)
print("Original array:\n", a)

a[:, [0, 2]] = a[:, [2, 0]]
print("After swapping:\n",a)

print("\n")

#Using np.copy()
a = np.arange(12).reshape(4, 3)
print("Original array:\n", a)

res = a.copy()
res[:, [0, 2]] = res[:, [2, 0]]
print("After swapping:\n", res)

print("\n")

arr = np.arange(16).reshape(4, 4)
print("My array before swapping:\n", arr)
arr[:, [0,3]] = arr[:, [3, 0]]
print("After swapping:\n", arr)
```

```python
#Joining NumPy Array
#Using numpy.concatenate()
a = np.array([1, 2])
b = np.array([3, 4])

res = np.concatenate((a, b))
print(res)
print("\n")

#Using numpy.hstack() / numpy.vstack() / numpy.dstack()
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

res_h = np.hstack((a, b))
print("np.hstack: ", res_h)

res_v = np.vstack((a, b))
print("np.vstack: ", res_v)

res_d = np.dstack((a,b))
print("np.dstack: ", res_d)
print("\n")

#Using numpy.stack()
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])

res = np.stack((a, b), axis=1)
res1 = np.stack((a, b), axis=0)
print("Vertical stacked array:\n",res)
print("Horizontal stacked array:\n", res1)
print("\n")

#Using numpy.block()
b1 = np.array([[1, 1], [1, 1]])
b2 = np.array([[2, 2, 2], [2, 2, 2]])
b3 = np.array([[3, 3], [3, 3], [3, 3]])
b4 = np.array([[4, 4, 4], [4, 4, 4], [4, 4, 4]])

res = np.block([
    [b1, b2],
    [b3, b4]
])

print(res)
print("\n")

class1 = np.array([[1,1,1],[2, 2, 2]])
class2 = np.array([[3, 3, 3, 3], [4, 4, 4, 4]])
class3 = np.array([[5, 5, 5], [6, 6, 6]])
class4 = np.array([[7,7,7,7], [8, 8, 8, 8]])

combined = np.block([[class1, class2],
                    [class3, class4]])
print(combined)
```

```python
#Combining a One and a Two-Dimensional NumPy Array
num1 = np.array([1, 2])
num2 = np.array([[10, 20],
                 [30, 40]])
                   
for a, b in np.nditer([num1, num2]):
    print(a, ":", b)
print("\n")

#Example 1: In this example, we combine a 1-D array of 5 elements with a 2-D array (2×5) 
#and display them together.
num1 = np.arange(5)
print("1D array:")
print(num1)

num2 = np.arange(10).reshape(2, 5)
print("\n2D array:")
print(num2)

# Combine 1-D and 2-D arrays
for a, b in np.nditer([num1, num2]):
    print("%d:%d" % (a, b))

print("\n")

a = np.arange(4)
b = np.arange(12).reshape(3, 4)
print("One-dimensional array:\n", a)
print("two-dimensional array:\n", b)
print("My array iteration:")
for x,y in np.nditer([a, b]):
    print(x, ":", y)
```
